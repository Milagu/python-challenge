```python
# PyBoss Homework #1 / Jamuna Prakash - 02/01/2019
# Import Dependencies
import pandas as pd      # Module to work with tabulated data in dataframes and lists
import csv               # Module to work with csv files
import re                # Module to work with Regular expressions
```

```python
# Make a reference to the employeedata.csv file path
csvpath = ("Resources/employee_data.csv")
```

```python
# Get the data onto a dataframe object
emp_input_df = pd.read_csv(csvpath, encoding = "UTF8")
```

```python
# Take a peek into the data
emp_input_df.head()
```

```python
# Split the Name column into two parts as first name and last name
new_df = emp_input_df["Name"].str.split(pat = " ", n = 1, expand = True)
```

```python
# Now, let's start being the BOSS..!
#1* The `Name` column should be listed as `First Name` and `Last Name` columns.

# First store the split strings in to two lists / dataframes
emp_input_df["First Name"] = new_df[0]
emp_input_df["Last Name"] = new_df[1]

emp_input_df.drop(["Name"], axis = 1)

# The following statement is important. This will ensure that the changes to the dataframe aren't lost 
emp_input_df = emp_input_df

# Take a peek into the data
emp_input_df.head()
```

```python
#2* The `DOB` data should be re-written into `MM/DD/YYYY` format.
#First convert the date in the dataframe from string format to date format
emp_input_df["DOB"] = pd.to_datetime(emp_input_df["DOB"]).dt.strftime('%m/%d/%Y')
```

```python
# Take a peek into the data
emp_input_df.head()
```

```python
#3* The `SSN` data should be re-written such that the first five numbers are hidden from view.
# Following code extracts first five digits of SSN using regular expression and replace with *s
emp_input_df["SSN"] = emp_input_df["SSN"].replace('^\d{3}[-]\d{2}[-]', '***-**-', regex = True) # this substitutes the first 5 digits of SSN with *s
# Take a peek into the data
emp_input_df.head()
```

```python
#4* The `State` data should be re-written as simple two-letter abbreviations.
# Create a dictionary to hold the State names as keys and abbreviations as values
#state abbr dictionary
us_state_ab_dict = {
    'Alabama': 'AL',
    'Alaska': 'AK',
    'Arizona': 'AZ',
    'Arkansas': 'AR',
    'California': 'CA',
    'Colorado': 'CO',
    'Connecticut': 'CT',
    'Delaware': 'DE',
    'Florida': 'FL',
    'Georgia': 'GA',
    'Hawaii': 'HI',
    'Idaho': 'ID',
    'Illinois': 'IL',
    'Indiana': 'IN',
    'Iowa': 'IA',
    'Kansas': 'KS',
    'Kentucky': 'KY',
    'Louisiana': 'LA',
    'Maine': 'ME',
    'Maryland': 'MD',
    'Massachusetts': 'MA',
    'Michigan': 'MI',
    'Minnesota': 'MN',
    'Mississippi': 'MS',
    'Missouri': 'MO',
    'Montana': 'MT',
    'Nebraska': 'NE',
    'Nevada': 'NV',
    'New Hampshire': 'NH',
    'New Jersey': 'NJ',
    'New Mexico': 'NM',
    'New York': 'NY',
    'North Carolina': 'NC',
    'North Dakota': 'ND',
    'Ohio': 'OH',
    'Oklahoma': 'OK',
    'Oregon': 'OR',
    'Pennsylvania': 'PA',
    'Rhode Island': 'RI',
    'South Carolina': 'SC',
    'South Dakota': 'SD',
    'Tennessee': 'TN',
    'Texas': 'TX',
    'Utah': 'UT',
    'Vermont': 'VT',
    'Virginia': 'VA',
    'Washington': 'WA',
    'West Virginia': 'WV',
    'Wisconsin': 'WI',
    'Wyoming': 'WY',
}
emp_input_df["State"].replace(us_state_ab_dict, inplace = True)
# Take a peek into the data
emp_input_df.head()
```

```python
# Now we will reorder the columns in the dataframe
#cols = list(emp_input_df.columns.values)
#print(cols)
emp_input_df = emp_input_df[['Emp ID', 'First Name', 'Last Name', 'DOB', 'SSN', 'State']]
# Take a peak into the data
emp_input_df.head()
```

```python
# Let's write the final output to a csv file
emp_input_df.to_csv("Output/emp_output.csv", index=False, header=True)
```
